**Enviando o Dataframe para o Banco de Dados e Realacionamento**

```python
import pandas as pd
import sqlalchemy
```

```python
arquivo = ('http://api.sidra.ibge.gov.br/values/t/418/n103/all/v/all/p/all/c216/0,3808,3811,3815,3818,3821,3825,3829,3836,3840,3843,3846,3850,3853,3856,3859,3864,104327,104330/c201/6196/d/v180%203?formato=json')
```

```python
df = pd.read_json(arquivo)
```

```python
df.head()
```

```python
#engine de conexão

engine = sqlalchemy.create_engine('mysql+pymysql://root:Salmos4;1@localhost:3306/brincar')
```

```python
df.to_sql(
    name='tb_medelete',
    con= engine,
    index = False,
    if_exists = 'relase'
)
```
