# Missing Values

* Missing values, são valores faltantes em colunas, esses podem ser oriundos em cargas de dados, falhas em crawlers ou ate mesmo corrupção de dados
* Missing values podem ser um problema em varias situações, como por exemplo, algortimo de machine learning que não trabalha bem com dados faltantes.
* Estes também pode atrapalhar resultado de análises  
* Vamos aprender como encontrar missing values na base de dados e como manipular esses valores

```python
import pandas as pd
arquivo = 'kc_house_data.csv'
dataset = pd.read_csv(arquivo, sep=',', header=0)
```

```python
#Consultando as linhas com valores faltantes.
dataset.isnull().sum()
```

```python
#Removendo as linha com valores faltantes com o metodo --dropna(inplace=True)
dataset.dropna(inplace=True)
```

```python
#Removendo as linha com valores faltantes em todas as colunas com o metodo --dropna(how='all', inplace=True)
dataset.dropna(how='all', inplace=True)
```

```python
#Adicionando um valor medio, os dados faltantes com o metodo fillna()
dataset['floors'].fillna(dataset['floors'].mean(), inplace=True)
```

```python
#Adicionando o valor qualquer com o metodo fillna()
dataset['bedrooms'].fillna(1, inplace=True)
```

# Visualização de dados
