# Trabalhando com visualização de dados estatisticos

```python
import pandas as pd
import matplotlib as mpl
```

```python
arquivo = 'kc_house_data.csv'
dataset = pd.read_csv(arquivo, sep=',')
```

```python
#Plotando um histograma com a tabela price
dataset['price'].hist(bins=30, color='red')
```

```python
#Plotando um histograma das colunas bedrooms e bathrooms lado a lado na cor verde

dataset[['bedrooms','bathrooms']].hist(bins=30, alpha=0.9, color='Green')
```

# Estatistica descritiva

```python
#Pegando a o valor medio de bedrooms
dataset['bedrooms'].mean()
```

```python
#Pegando a o valor maximo de bedrooms
dataset['bedrooms'].max()
```

```python
#Pegando a o valor minimo de bedrooms
dataset['bedrooms'].min()
```

```python
#Pegando a o valor desvio padrão de bedrooms
dataset['bedrooms'].std()
```

```python
#vendo a assimetria da coluna bedrooms
dataset['bedrooms'].skew()
```

```python
#vendo a assimetria dos dados das colunas de uma dataset
dataset.skew()
```

# Extraindo Insights

* Trabalhando com o grafico de boxplot. Os grafico de boxplot são excelentes para identificação de anomalias nos dados

```python
#Plotando um grafico de boxplot na coluna bedrooms
#Posso mudar o style do grafico usando a biblioteca matplotlib.
%matplotlib notebook
mpl.style.use('ggplot')
dataset.boxplot(column='bedrooms')
```

```python
#Vendo todo os stylos de grafico.
mpl.style.available
```

```python
#plotando no grafico de boxplot o preço por numero de quartos

dataset.boxplot(column='price', by='bedrooms')
```

# Trabalhando com correlação de dados

```python
#Correlação de todos os dados do dataset
dataset.corr()
```

```python
#Imprimindo a correlação de todo o dataset, esse metodo leva em consideração as positiva e negativas
dataset.corr('spearman')
```

```python
#Imprimindo a correlação de algumas colunas
dataset[['bedrooms','bathrooms', 'sqft_living','floors','waterfront','grade','price']].corr()
```

```python
#Plotando a correlação de algumas colunas
dataset[['sqft_living','waterfront','grade','price']].corr().plot()
```
