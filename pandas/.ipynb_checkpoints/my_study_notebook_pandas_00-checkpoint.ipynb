```python
import pandas as pd
```

```python
#Importando a base de dados
#Lendo a base com pandas
#   --sep=','-- indica o tipo de separação dos dados 
#   --header=0-- diz onde está os nomes das colunas

arquivo = 'kc_house_data.csv'  
dataset = pd.read_csv(arquivo, sep=',', header=0) 
```

```python
#Verificando o tipo da vareavel
type(dataset)
```

```python
#Imprimindo as 5 primeiras linhas do Dataframe
#   --.head()-- caso queira ver mais de 5 linhas basta passar o numero de linha por parametro .head(100)

dataset.head()
```

```python
#Carregando a base de dados e indexando a base pela coluna date
#   --index_col=''-- é usado para indexar o dataframe por alguma coluna
dataset = pd.read_csv(arquivo, sep=',', index_col='date')
```

```python
#Carregandoo a base de dados e selecionando as colunas que serão usada
#   --usecols=[]-- é usado para selecionar as colunas que serão usadas
dataset = pd.read_csv(arquivo, sep=',', usecols=['id','date','price','bedrooms'])
```

```python
#Imprimindo a lista de colunas com o metodo .columns
dataset.columns
```

```python
#O metodo count() mostra a quantidade de linhas que tem em cada coluna do dataframe
#   --count()--
dataset.count()
```

```python
#O metodo describe() exibe informações estatisticas da base de dados
#Informações como: media,desvio padrão, valor minimo e maixo das colunas
#   --.describe()--
dataset.describe()
```

```python
#Esse metodo imprime as ultimas linhas do datafreme,
#   --.tail()-- é possivel usar parametro para especificar a quantidade de linhas .tail(100)
dataset.tail()
```

```python
#Imprime uma amostra aleatoria do dataset
#  --ssample(5)--  para determinar a quantidade da amostra basta passa-la o numero por parametro
dataset.sample(5)
```

```python
#Retorna a quantidade de linhas e dolunas do datafreme em formato de tupla
dataset.shape
```

```python
#O metodo .info() imprime informações sobre colunas e uso de memoria
#   informações como:
#   -Quantidade de linhas de cada colunas
#   -Se existe algum valor nulo
#   -Tipo dos dados de cada coluna
#   -Quantidade de memoria que o Dataframe está utilizando
dataset.info()
```

```python
#Importando o pandas_profiling
import pandas_profiling
```

```python
#Carregando a base de dados
arquivo = 'kc_house_data.csv'
```

```python
#Lendo a base dados
df = pd.read_csv(arquivo, sep=',', header=0)
```

```python
#Ulrilizando o pandas profile
pandas_profiling.ProfileReport(df)
```

```python
#Carregando o resultado do relatorio na vareavel profile
profile = pandas_profiling.ProfileReport(df)
```

```python
#Exportando o 
profile.to_file("report.html")
```
