# Trabalhando com Grandes Arquivos

```python
import pandas as pd
arquivo = 'kc_house_data.csv'
```

```python
#Lendo as 5 primeiras linhas do arquivo.
#   --nrows='n'--
dataset = pd.read_csv(arquivo, sep=',')
```

```python
#O dataframe dataset, tem somente 5 linhas que foram carregadas pelo metodo 'nrows'
dataset.head(10)
```

```python
#Ultilizando o parametro chunksize podemos definir um bloco de dados com uma determinada quantidade de linhas
chunk = pd.read_csv(arquivo, chunksize=10000)
```

```python
type(chunk)
```

```python
#Imprimindo a quantidade de linhas em cada bloco
for parte in chunk:
    print (len(parte))
```

```python
chunk = pd.read_csv(arquivo, chunksize=10000)
```

```python
#Nesse bloco de condigo, crio uma lista que irá receber os elemento de 'bedrooms' e muiltiplico por 2,
#adiciono com o metodo --append-- no final da lista.
#Na ultima linha crio uma nova coluna chamada 'bedrooms_size' e com o metodo --concat()-- adiciono os dados
#de lista em a nova coluna 'bedrooms' do meu dataset.
lista = []
for partes in chunk:
    lista.append(partes['bedrooms'] * 2)
    
dataset['bedrooms_size'] = pd.concat(lista)
```

# Trabalhando com os tipos de dados adequados

* Atenção para os tipos de object
* Dados que são categoricos podem receber o tipo de dados caregory

```python
#Carregando uma base da web
df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
```

```python
#Verificando o tamanha da memoria do dataframe com o metodo --.info()--
df.info()
```

```python
#convetendo os tipo de dados
df.Sex = df.Sex.astype('category')
df.Embarked = df.Embarked.astype('category')
df.Survived = df.Survived.astype('category')
df.Pclass = df.Pclass.astype('category')
df.PassengerId = df.PassengerId.astype('int32')
df.Parch = df.Parch.astype('int32')
df.SibSp = df.SibSp.astype('int32')
```

```python
#Vendo as mudanças no consumo de memoria do Dataframe
df.info()
```

```python
#Convertendo os valores na leitura do datase
data = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(data, dtype = {"Embarked" : "category", "Survived" : "category", "Parch" : "int32"})
```

```python
#Vendo as mudanças com o df.info()
df.info()
```

# Consultando um Dataset

* Podemos fazer conultas em Dataframe isso se assemelha a linha SQL.
* As consultas podem ser feitas por instruções de agrupamentos

```python
#Contar a quantidade de valores unicos
pd.value_counts(dataset['bedrooms'])
```

```python
#Consultando os número de imoveis que tem a quantidade de quartos iguais a 3
#Usando o metodo --.loc[lista]---, O metodo loc recerbe uuma lista como parametro
dataset.loc[dataset['bedrooms']==3]
```
