# Alterando um Dataframe

```python
#Criando uma nova coluna e adicionando valores com uma calculo matematico
dataset['size'] = (dataset['bedrooms'] * 20)
```

```python
#Visualizando o contÃºdo de uma coluna
dataset['size'].head()
```

```python
#Criando uma funÃ§Ã£o para processamento de dados
def categoriza(x):
    if x >= 80:
        return 'Big'
    elif x >= 60:
        return 'Medium'
    elif x >= 40:
        return 'small'
```

```python
#Usando uma funÃ§Ã£o para categorizar os imoveis
#Criando uma lista = aplincando a funÃ§Ã£o na coluna size
dataset['cat_size'] = (dataset['size'].apply(categoriza))
```

```python
#Mostrando a coluna cat_size
dataset['cat_size']
```

```python
#Verificando a quantidade de valores de cada categoria
#Usando o metodo --pd.value_counts(lista)
pd.value_counts(dataset['cat_size'])
```

```python
#Apagando dados do dataset
#axis=1 apaga uma coluna
#axis=0 apaga uma linha
#inplace=True salva as alteraÃ§Ãµes em memoria
#inplace=False nÃ£o salva as alteraÃ§Ãµes em memoria

dataset.drop(['cat_size'], axis=1, inplace=True)
```

```python
#Verificando a coluna cat_size
dataset.head()
```

```python
#Apagando linhas como especificaÃ§Ãµes logicas
dataset.drop(dataset[dataset.bedrooms==0].index, inplace=True)
```

```python
#O codigo a baixo mostra que nao tem nem uma valor 0 na coluna bedrooms
pd.value_counts(dataset['bedrooms']==0)
```

```python
#Apagando todos os dados que tem a mais de 30 bedrooms com o metodo --.drop()
dataset.drop(dataset[dataset.bedrooms>30].index, inplace=True)
```

# Percorrendo linhas 

O iteroows serve para imprimir as linhas de um DataFrame

```python
#Imprimindo a primeira linha do objeto iterator
for indice, linha in dataset.head(10).iterrows():
    print(indice, linha)
```

```python
#Percorrendo o dataframe e acessando colunas por nomes
for indice, linha in dataset.head(10).iterrows():
    print(indice, linha['bedrooms'], linha['floors'], linha['price'])
```

Atualizando um dataframe ao percorrer linha a linha

```python
#imprimindo os 5 primeiros valores de preÃ§os antes de atualizar
dataset.price.head(5)
```

```python
#Percorrendo e atualizando as linha de um dataframe. 
#Atualizando o valor da coluna price, multiplicando o seu valor por 2
#Ã© preciso usar o mÃ©todo at() 
for indice, linha in dataset.iterrows():
    dataset.at[indice, 'price'] = linha['price'] * 2
```

```python
dataset.price.head()
```

```python
#Percorrendo o dataframe usando itertuples()
for linha in dataset.head().itertuples():
    print(linha)
```

```python
#Percorrendo o dataframe usando itertuples()
#Percorrendo linha especificas
for linha in dataset.head().itertuples():
    print(linha.id, linha.date, linha.price)
```
