```python
# Importando o pyplot
from matplotlib import pyplot as plt
```

O codigo abaico define que os grafico serão fixos na tela do jupyter notebook

# Graficos de linhas

* graficos normalmente utilizado para visualização de dados em uma linha do tempo
* Normalmente utilizado com o eixo X relacionado a dados temporais e eixo Y a dados quantitativo
* Cuidado ao plotar vairas linhas, pois a leitura pode ficar prejudicada

```python
# Usando o grafico dinaminco com o %mtplotlib notebook

%matplotlib notebook
#Eixo_X, Eixo_Y
plt.plot([1,2,3,4],[1,4,9,16])
plt.show()
```

```python
# Usando o grafico fixo com o %mtplotlib inline

%matplotlib inline
#Eixo_X, Eixo_Y
plt.plot([1,2,3,4],[1,4,9,16])
plt.show()
```

**Definição de label aos eixos**

```python
%matplotlib notebook
import matplotlib.pyplot as plt
plt.plot([1,2,3,4], [1,4,9,16])
plt.ylabel('Eixo Y')
plt.xlabel('Eixo X')
plt.show()
```

**Alterando a forma dos pontos**

```python
%matplotlib notebook
plt.plot([1,2,3,4], [1,4,9,16],'o')
plt.show()
```
