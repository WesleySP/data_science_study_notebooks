# Graficos de linhas

* graficos normalmente utilizado para visualização de dados em uma linha do tempo
* Normalmente utilizado com o eixo X relacionado a dados temporais e eixo Y a dados quantitativo
* Cuidado ao plotar vairas linhas, pois a leitura pode ficar prejudicada

```python
# Usando o grafico dinaminco com o %mtplotlib notebook

%matplotlib notebook
#Eixo_X, Eixo_Y
plt.plot([1,2,3,4],[1,4,9,16])
plt.show()
```

```python
# Usando o grafico fixo com o %mtplotlib inline

%matplotlib inline
#Eixo_X, Eixo_Y
plt.plot([1,2,3,4],[1,4,9,16])
plt.show()
```

# Plotando dados categoricos

* Trabalhando com gráficos de barras.
* Ideal para qunado estamos trabalhaco com **comparação de valores.**
* Normalmente quer se visualizar frequência ou percentual de valores.

**Plotando gráficos de barras verticais ou gráficos de colunas**

```python
%matplotlib inline
grupos = ['Produto A', 'Produto B', 'Produto C']
valores = [1, 10, 100]

plt.bar(grupos, valores)

plt.show
```

**Plotando grafico de barras juntas**

```python
# Importando o numpy
import numpy as np
```

```python
%matplotlib inline

# Quantidade de valores para o Produto A
valores_produto_A = [6, 7, 8, 4, 4]

# Quantidade de valores para o Produto B
valores_produto_B = [3, 12, 3, 4.1, 6]

# Criando o eixo x para os produtos A e produtos B com separação de 0.25 entre as barras
x1 = np.arange(len(valores_produto_A))
x2 = [x + 0.25 for x in x1]

#Plotando as barras
plt.bar(x1, valores_produto_A, width=0.25, label = "Produto A", color = 'b')
plt.bar(x2, valores_produto_B, width=0.25, label = "Produto B", color = 'y')

meses = ['Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro']
plt.xticks([x + 0.25 for x in range(len(valores_produto_A))], meses)

plt.legend()

plt.title('Quantidade de Vendas')
plt.show()
```

**Graficos de barras horizontais**


* Grafico muito utilizado qunado queremos fazer comparações
* Muito usado em capanha de pesquisas eleitorais

```python
%matplotlib inline
grupos = ['Produto A', 'Produto B', 'Produto C']
valores = [1, 10, 100]

plt.barh(grupos, valores)
plt.barh(grupos, valores)
plt.yticks(rotation=45)

plt.show
```

# Grafico Setorial (grafico de pizza)

```python
%matplotlib inline

vendas = [3000, 2300, 1000, 500]
labels = ['E-commerce', 'Loja Física', 'e-mail', 'Marketplace']

plt.pie(vendas, labels=labels)
plt.show()
```

# Customizando gráficos

* Adicionando informações de percetual, sobras, separação das partes e legendas

```python
%matplotlib inline

vendas = [3000, 2300, 1000, 500]
labels = ['E-commerce', 'Loja Física', 'e-mail', 'Marketplace']

#define o nível de separação entras as partes, ordem dos vetores representa as partes
explode = (0.1, 0, 0, 0)

plt.pie(vendas, labels=labels, autopct='%1.1f%%', shadow=True, explode=explode)
plt.legend(labels, loc=2)

#define que o grafico será plotado em circulo
plt.axis('equal')

plt.show()
```

# Criando Subplot

* Para criar subplots defina o número de linhas de gráficos e número de colunas(quantidade de gráfico em cada linha)

```python
plt.subplots(nrows= 1, ncols = 2)
```

**1 linhas com 3 colunas**

```python
plt.subplots(nrows= 1, ncols = 3)
```

**2 linhas 2 colunas**

```python
plt.subplots(nrows=2, ncols=2)
```

**Para subplots usa-se o método subplot (atenção para o nome correto no singular)**

```python
x = [1, 2, 3, 4]
y = [10, 20, 25, 40]

plt.subplot(1,2,1)
plt.plot(x,y, 'o')

plt.subplot(1,2,2)
plt.plot(x,y, color='r')
```

**Criação de instancias de classes figure**

* Aumenta o controle do nivel de customização das funções do Matplotlib
* Gera menos código devido a orientação a objetos.

```python
#Criando um objeto do tipo figure
fig = plt.figure()

#Define o tamanho em porcentagem do grafico.
#Altura e largura do eixo na figura
axes_a = fig.add_axes([0.1,0.1,0.8,0.8])

#plotando
axes_a.plot(x,y)
```

**Uma figura com dois gráficos de proporções diferentes**

```python
#Criando um objeto do tipo figure
fig = plt.figure()

#Define o tamanho em porcentagem do grafico.
#Altura e largura do eixo na figura
axes_a = fig.add_axes([0.1,0.1,0.8,0.8])
axes_b = fig.add_axes([0.2,0.5,0.3,0.3])

#Plota a figura com dois eixos.
#O "r--" significa linha tracejada em vermelho(r)

axes_a.plot(x,y)
axes_b.plot(x,y,'r--')
```
